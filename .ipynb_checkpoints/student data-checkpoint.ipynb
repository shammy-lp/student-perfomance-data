```python
if df.isnull().any().any():
    print("\nThere are missing values in the DataFrame.")
else:
    print("\nNo missing values in the DataFrame.")
```

```python
df=pd.DataFrame(df)
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df = pd.DataFrame(df)
columns_to_drop = ("Major Code",
                  "Spec",
                  "Birth Date",
                  "Actual Graduation Date")
columns_to_drop = [col.strip() for col in columns_to_drop]
df = df.drop(columns=columns_to_drop, errors='ignore')
print("\nDataFrame after dropping specified columns:")
print(df)
```

```python
df.head()
```

```python
# Check for duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Remove duplicate rows
df.drop_duplicates(inplace=True)
```
